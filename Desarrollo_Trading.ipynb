```python
def round_time(dt=None, date_delta=datetime.timedelta(minutes=1), to='average'):
    """
    Round a datetime object to a multiple of a timedelta
    dt : datetime.datetime object, default now.
    dateDelta : timedelta object, we round to a multiple of this, default 1 minute.
    from:  http://stackoverflow.com/questions/3463930/how-to-round-the-minute-of-a-datetime-object-python
    """
    #roud_time (redondea una fecha o una hora, al dato anterior múltiplo de 5)
    round_to = date_delta.total_seconds()
    if dt is None:
        dt = datetime.now()
    seconds = (dt - dt.min).seconds

    if seconds % round_to == 0:
        rounding = (seconds + round_to / 2) // round_to * round_to
    else:
        if to == 'up':
            # // is a floor division, not a comment on following line (like in javascript):
            rounding = (seconds + round_to) // round_to * round_to
        elif to == 'down':
            rounding = seconds // round_to * round_to
        else:
            rounding = (seconds + round_to / 2) // round_to * round_to

    return dt + datetime.timedelta(0, rounding - seconds, -dt.microsecond)
```

```python
def data_constructor(df,  time_window):
    
    start_rounded =round_time(datetime.datetime(df['Dates'][0].year,df['Dates'][0].month,df['Dates'][0].day,
                                                df['Dates'][0].hour, df['Dates'][0].minute, df['Dates'][0].second),
                                                date_delta=datetime.timedelta(minutes=5),to='down')

    
    high_ask = []
    low_ask = []
    high_bid = []
    low_bid = []
    close_ask = []
    close_bid = []
    midprice = []
    dates = []

    count = 0

    inicio = start_rounded
    fin = start_rounded + timedelta(minutes=time_window)
    
    
    while inicio <= df['Dates'][len(df)-1]:

        rango = df.loc[(df['Dates'] >= inicio) & (df['Dates'] < fin)]

        if len(rango) > 0:

            high_ask.append(max(rango['Ask']))
            low_ask.append(min(rango['Ask']))
            high_bid.append(max(rango['Bid']))
            low_bid.append(min(rango['Bid']))
            a = rango['Ask'].tolist()
            b = rango['Bid'].tolist()
            close_ask.append(a[-1])
            close_bid.append(b[-1])


        else:
            #En caso de que no haya datos en esa ventana de tiempo
            
            high_ask.append(high_ask[count - 1])
            low_ask.append(low_ask[count - 1])
            high_bid.append(high_bid[count - 1])
            low_bid.append(low_bid[count - 1])
            close_ask.append(close_ask[count-1])
            close_bid.append(close_bid[count-1])

        dates.append(inicio)
        midprice.append((close_ask[count]+close_bid[count]) / 2)
        inicio = inicio + timedelta(minutes=time_window)
        fin = fin + timedelta(minutes=time_window)
        count += 1


    d = {'Dates': dates, 'Close Ask': close_ask, 'Close Bid': close_bid, 'High Ask': high_ask,
                'High Bid': high_bid, 'Low Ask': low_ask, 'Low Bid': low_bid, 'MidPrice': midprice}
    bd = pd.DataFrame(data=d)

    return bd
```

```python
bd5 = data_constructor(df, time_window=5)
bd20 = data_constructor(df, time_window=20)
bd50 = data_constructor(df, time_window=50)
```

```python
#Calculo de media móvil simple
def MA(df,n):
    MA = pd.Series(pd.Series.rolling(df['MidPrice'], n).mean(), name = 'MA_'+ str(n))
    df = df.join(MA)
    return df

#Cuando la media móvil simple más rápida cruza por encima de la más lenta, daría una señal de compra
#Cuando la media móvil simple más rápida cruza por debajo de la más lenta, daría una señal de venta
```

```python
#Calculo de media móvil exponencial
def EMA(df,n):
    EMA = pd.Series(pd.Series.ewm(df['MidPrice'], span = n, min_periods = n-1, adjust = False).mean(), name = 'EMA_' + str(n))
    df = df.join(EMA)
    return df
```

```python
#Calculo de momentum
def MOM(df, n):
    M = pd.Series(df['MidPrice'].diff(n), name = 'Momentum_' + str(n))
    df = df.join(M)
    return df
```

```python
#Calculo de Rate of Change (ROC)
def ROC(df, n):
    M = df['MidPrice'].diff(n-1)
    N = df['MidPrice'].shift(n-1)
    ROC = pd.Series(M/N, name = 'ROC_' + str(n))
    df = df.join(ROC)
    return df
#Mide el cambio porcentual entre el precio actual y el precio pasado del período n.
```

```python
#Calculo de utilidad Bruta
sell = []
buy = []
datesell = []
datebuy = []
cl = t.Close
da = t.Dates
bs = t.BuyOrSell
bs = bs.replace('vender',"0")
bs = bs.replace('comprar',"1")

for n in range(len(t.Close)):
        if bs[n] == '0':
            sell.append(cl[n])
            datesell.append(da[n])
        elif bs[n] == '1':
            buy.append(cl[n])
            datebuy.append(da[n])
            
ventas = sum(sell)
compras = sum(buy)

MargenUtilidadBruta = ventas - compras
MargenProcentual = (MargenUtilidadBruta / ventas) * 100

S = {'DateS': datesell, 'Sell': sell}
s = pd.DataFrame(data=S)

B = {'DateB': datebuy, 'Buy': buy}
b = pd.DataFrame(data=B)

print('Ventas', '=', ventas)
print('Compras', '=', compras)
print('MargenUtilidadBruta', '=', MargenUtilidadBruta)
print('MargenPorcentual', '=', MargenProcentual)
```

```python
#Calculo de utilidad 
utilidad_tot = 0
perdidas = 0
ganancias = 0
acum = []
for i in range(len(t)-1):
    if t['BuyOrSell'][i] == 'comprar':
        index = bd20[bd20['Dates'] == t['Dates'][i]].index[0]
        utilidad_op = bd20['Close Bid'][index+1] - bd20['Close Ask'][index]

        
    elif t['BuyOrSell'][i] == 'vender':
        index = bd20[bd20['Dates'] == t['Dates'][i]].index[0]
        utilidad_op = -(bd20['Close Ask'][index+1] - bd20['Close Bid'][index])

    if utilidad_op < 0:
        perdidas = perdidas + abs(utilidad_op)
    elif utilidad_op >0:
        ganancias = ganancias + abs(utilidad_op)
        
    utilidad_tot = utilidad_tot + utilidad_op
    acum.append(utilidad_tot + utilidad_op+1000)
    
    
print('La utilidad total es de ', round(utilidad_tot,2), ' pesos por cada dólar invertido')
print('Ganancia bruta: ', ganancias)
print('Pérdida bruta: ', perdidas)
```

```python
#Medidas de desempeño

#Calculo profit factor.
ProfitFactor = ganancias/perdidas
print('Profit Factor', '=', ProfitFactor)
#Un Profit Factor por  encima de 2 es sobresaliente,cuanto mayor es el número, mejor. 
#Un Profit Factor de beneficio de 3 significa que sus ganancias netas fueron 3 veces mayores que  sus pérdidas netas.
```
